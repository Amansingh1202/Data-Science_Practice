```python
data.loc[data['sales']=='sales','sales'] = 1
data.loc[data['sales']=='accounting','sales'] = 2
data.loc[data['sales']=='hr','sales'] = 3
data.loc[data['sales']=='technical','sales'] = 4
data.loc[data['sales']=='support','sales'] = 5
data.loc[data['sales']=='management','sales'] = 6
data.loc[data['sales']=='IT','sales'] = 7
data.loc[data['sales']=='product_mng','sales'] = 8
data.loc[data['sales']=='marketing','sales'] = 9
data.loc[data['sales']=='RandD','sales'] = 10

data.loc[data['salary']=='low','salary'] = 1
data.loc[data['salary']=='medium','salary'] = 2
data.loc[data['salary']=='high','salary'] = 3
```

```python
df = data.copy()
y = df['left']
x = df.drop(columns = ['left'])
```

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn import svm
clf_svm = svm.SVC()
clf_svm.fit(x_train,y_train)
```

```python
clf_svm.predict([x_test.iloc[3]])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rfc = RandomForestClassifier(n_estimators = 100)
clf_rfc.fit(x_train,y_train)
```

```python
clf_rfc.predict([x_test.iloc[1]])
```

```python
from sklearn.metrics import f1_score
print(clf_svm.score(x_test,y_test))
print(f1_score(y_test,clf_svm.predict(x_test),average=None,labels=[0,1]))
```

```python
print(clf_rfc.score(x_test,y_test))
print(f1_score(y_test,clf_rfc.predict(x_test),average=None,labels=[0,1]))
```

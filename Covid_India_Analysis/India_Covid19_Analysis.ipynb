```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
age_group = pd.read_csv('AgeGroupDetails.csv')
corona_detail = pd.read_csv('covid_19_india.csv')
hospital_beds = pd.read_csv('HospitalBedsIndia.csv')
statewise_data = pd.read_csv('StatewiseTestingDetails.csv')
corona_detail_grouped = corona_detail.groupby('State/UnionTerritory')
corona_detail.head()
```

```python
for name,group in corona_detail_grouped:
    group.reset_index(drop = True,inplace=True)
    g = sns.catplot(x='Date',y='Confirmed',aspect = 5,data=group,kind='bar',legend = 'brief').set(title = name)
    g.set_xticklabels(rotation=30)
```

```python
for name,group in corona_detail_grouped:
    group.reset_index(drop = True,inplace=True)
    g = sns.lineplot(x='Confirmed',y='Cured',legend = 'brief',data=group).set(title = name)
    plt.show()
```

```python
for name,group in corona_detail_grouped:
    group.reset_index(drop = True,inplace=True)
    g = sns.lineplot(x='Confirmed',y='Deaths',legend = 'brief',data=group).set(title = name)
    plt.show()
```
